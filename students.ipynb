```python
import pandas as pd

# Create sample student data
data = {
    'StudentID': range(1, 11),
    'Name': ['Anelisa', 'Xhanti', 'Cecelia', 'David', 'Inga', 'Fundi', 'Grace', 'Lungi', 'Ben', 'Julia'],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F', 'F', 'M', 'F'],
    'Age': [16, 17, 16, 18, 17, 16, 15, 16, 17, 18],
    'Grade': [10, 11, 10, 12, 11, 10, 9, 10, 11, 12],
    'MathScore': [88, 75, 82, 90, 70, 85, 78, 92, 73, 89],
    'ReadingScore': [90, 78, 85, 87, 75, 88, 82, 91, 76, 90],
    'WritingScore': [85, 72, 80, 88, 74, 84, 79, 93, 71, 88]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display first few rows
df.head()
```

```python
# Dataset Info
print(df.info())

# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# Summary statistics
print("\nDescriptive statistics:")
print(df.describe())
```

```python
# Average scores by gender
print("\nAverage scores by gender:")
print(df.groupby('Gender')[['MathScore', 'ReadingScore', 'WritingScore']].mean())

# Average scores by grade
print("\nAverage scores by grade:")
print(df.groupby('Grade')[['MathScore', 'ReadingScore', 'WritingScore']].mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='StudentID', y='MathScore', data=df, marker='o')
plt.title("Math Score by Student")
plt.xlabel("Student ID")
plt.ylabel("Math Score")
plt.show()
```

```python
plt.figure(figsize=(7, 4))
sns.barplot(x='Gender', y='WritingScore', data=df)
plt.title("Average Writing Score by Gender")
plt.ylabel("Writing Score")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['ReadingScore'], bins=10, kde=True)
plt.title("Distribution of Reading Scores")
plt.xlabel("Reading Score")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='MathScore', y='WritingScore', hue='Gender', data=df)
plt.title("Math Score vs. Writing Score")
plt.xlabel("Math Score")
plt.ylabel("Writing Score")
plt.legend(title="Gender")
plt.show()
```
